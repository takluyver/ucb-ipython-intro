A notebook can even contain figures:

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 10, 1000)
y = t
for omega in range(1,8):
    y = y * np.sin(omega * t)
    plt.plot(t, y)
```
